```python
import requests
```

```python
form_data = {'SearchType' : 'S',
'Lang' : 'TW',
'StartStation' : 'NanGang',
'EndStation' : 'ZuoYing',
'OutWardSearchDate' : '2021/05/20',
'OutWardSearchTime' : '16:30',
'ReturnSearchDate' : '2021/05/19',
'ReturnSearchTime' : '22:30'}
```

```python
res = requests.post("https://www.thsrc.com.tw/TimeTable/Search", data = form_data)
```

```python
import json
```

```python
data = json.loads(res.text)
```

```python
data
```

```python
clean = data['data']['DepartureTable']['TrainItem']
```

```python
print('車次', '出發', '抵達', '耗時')

for item in clean:
    print(item['TrainNumber'], item['DepartureTime'], item['DestinationTime'])
```
